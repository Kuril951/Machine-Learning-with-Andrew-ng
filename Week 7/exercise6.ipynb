```python
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio

%matplotlib inline
```

```python
data = sio.loadmat("./machine-learning-ex6/ex6/ex6data1.mat")

train_X = np.array(data['X'])
train_Y = np.array(data['y'])

pos = np.array([train_X[i] for i in range(len(train_X)) if train_Y[i] == 0])
neg = np.array([train_X[i] for i in range(len(train_X)) if train_Y[i] == 1])

print(train_X.shape, train_Y.shape)
```

```python
 def showData():
    plt.figure(figsize=(8, 6))
    plt.plot(pos[:,0], pos[:,1], 'yo')
    plt.plot(neg[:,0], neg[:,1], '+')
    
showData()
```

```python
def showLine(clf):
    showData()
    W = clf.coef_[0]
    b = clf.intercept_[0]
    
    x = np.arange(np.min(train_X[:,0]), np.max(train_X[:,0]), 0.05)
    y = -(W[0] * x + b) / W[1]
    plt.plot(x, y, '-')
    
def main1():
    
    # C = 1
    clf = svm.LinearSVC(C=1)
    clf.fit(train_X, train_Y.flatten())
    showLine(clf)
    
    # C = 100
    clf2 = svm.LinearSVC(C=100)
    clf2.fit(train_X, train_Y.flatten())
    showLine(clf2)
    
main1()
```

```python
data2 = sio.loadmat('./machine-learning-ex6/ex6/ex6data2.mat')

train_X = np.array(data2['X'])
train_Y = np.array(data2['y'])

pos = np.array([train_X[i] for i in range(len(train_X)) if train_Y[i] == 0])
neg = np.array([train_X[i] for i in range(len(train_X)) if train_Y[i] == 1])

print(train_X.shape)

def showData(pos, neg):
    plt.figure(figsize=(10, 8))
    plt.plot(pos[:,0], pos[:,1], 'yo')
    plt.plot(neg[:,0], neg[:,1], '+')

showData(pos, neg)
```

```python
def showLine(pos, neg, rbf):
    showData(pos, neg)
    x = np.arange(np.min(train_X[:,0]), np.max(train_X[:,0]), 0.02)
    y = np.arange(np.min(train_X[:,0:1]), np.max(train_X[:,0:1]), 0.02)
    XX, YY = np.meshgrid(x, y)
    
    ZZ = rbf.predict(np.c_[XX.ravel(), YY.ravel()])
    ZZ = ZZ.reshape(XX.shape)
    plt.contour(XX, YY, ZZ)
    
def main2():
    C = 1
    sigma = 0.1

    rbf = svm.SVC(kernel='rbf', C=C, gamma=(1/(2*np.square(sigma))))
    rbf.fit(train_X, train_Y.flatten())
    showLine(pos, neg, rbf)

main2()
```

```python
data3 = sio.loadmat('./machine-learning-ex6/ex6/ex6data3.mat')

train_X = np.array(data3['X'])
train_y = np.array(data3['y'])

cv_X = np.array(data3['Xval'])
cv_y = np.array(data3['yval'])

pos = np.array([train_X[i] for i in range(len(train_X)) if train_y[i] == 0])
neg = np.array([train_X[i] for i in range(len(train_X)) if train_y[i] == 1])
showData(pos, neg)
```

```python
def showLine(pos, neg, rbf):
    showData(pos, neg)
    x = np.linspace(np.min(train_X[:,0]), np.max(train_X[:,0]), 100)
    y = np.linspace(np.min(train_X[:,1]), np.max(train_X[:,1]), 100)
    XX, YY = np.meshgrid(x, y)
    ZZ = rbf.predict(np.c_[XX.ravel(), YY.ravel()])
    ZZ = ZZ.reshape(XX.shape)
    plt.contour(XX, YY, ZZ)
    
def main3():
    C = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
    sigma = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
    
    m = len(cv_X)
    
    accuracy = []
    for c in C:
        for sig in sigma:
            rbf = svm.SVC(kernel='rbf', C=c, gamma=(1/(2*np.square(sig))))
            rbf.fit(train_X, train_y.flatten())
            predict = rbf.predict(cv_X)
            accuracy.append([np.sum(np.equal(cv_y.flatten(), predict)) / m, c, sig])
            
    param = accuracy[np.argmax(accuracy, axis=0)[0]]
    print("C:", param[1], ", sigma:", param[2])
    rbf = svm.SVC(kernel='rbf', C=param[1], gamma=(1/(2*np.square(param[2]))))
    rbf.fit(train_X, train_y.flatten())
    
    showLine(pos, neg, rbf)
    
main3()
```

# Spam Classification

```python
import re
from nltk.stem import PorterStemmer

def readEmail(str):
    f = open(str, 'r')
    email = f.readlines()
    f.close()
    return email

def getVocabList():
    f = open('./machine-learning-ex6/ex6/vocab.txt')
    vocabList = f.readlines()
    f.close()
    n = len(vocabList)
    
    voca = {}
    num_to_voca = {}
    for i in range(n):
        idx, word = vocabList[i].split()[0], vocabList[i].split()[1]
        voca[word] = idx
        num_to_voca[int(idx)] = word
    return voca, num_to_voca

def processEmail(email):
    # lower case
    email_contents = ''.join(email).lower()
    
    # strip all HTML
    email_contents = re.sub('<[^<>]+>', ' ', email_contents)
    
    # handle Numbers
    email_contents = re.sub('[0-9]+', 'number', email_contents)
    
    # handle urls  http:// https://
    email_contents = re.sub('(http|https)://[^\s]*', 'httpaddr', email_contents)
    
    # handle Email addresses
    email_contents = re.sub('[^\s]+@[^\s]+', 'emailaddr', email_contents)
    
    # handle $ sign
    email_contents = re.sub('[$]+', 'dollar ', email_contents)
    
    # tokenize email
    email_contents = re.sub('[^a-zA-Z0-9]', ' ', email_contents).strip()
    email_contents = re.sub('[ ]+', ' ', email_contents)
    
    # Word Stemming
    word_list = email_contents.split()
    ps = PorterStemmer()
    for i in range(len(word_list)):
        word_list[i] = ps.stem(word_list[i])
    
    return word_list

def emailFeatures(word_indices, vocabList):
    x = np.zeros((len(vocabList)))
    
    word_to_number = []
    for i in range(len(word_indices)):
        if(word_indices[i] in vocabList):
            word_to_number.append(vocabList[word_indices[i]])
            x[int(vocabList[word_indices[i]]) - 1] = 1
    
    return word_to_number, x

def topRankedWords(vocabList, W):
    w = np.array([[i + 1, W[i]] for i in range(len(W))])
    w = np.array(sorted(w, key=lambda w:w[1], reverse=True))
    
    ranking = []
    for i in range(15):
        ranking.append(vocabList[int(w[i][0])])
    print(ranking)
    
def main2_1():
    email = readEmail('./machine-learning-ex6/ex6/emailSample1.txt')
    vocabList, num_to_voca_list = getVocabList()
    word_indices = processEmail(email)
    word_vector, features = emailFeatures(word_indices, vocabList)
    
    # Train Linear Svm for Spam Calssification
    train_data = sio.loadmat('./machine-learning-ex6/ex6/spamTrain.mat')
    train_X = train_data['X']
    train_y = train_data['y']
    
    m = len(train_X)
    C = 1e-1
    clf = svm.LinearSVC(C=C)
    clf.fit(train_X, train_y.flatten())
    predict = clf.predict(train_X).reshape((-1, 1))

    accuracy = np.mean(predict == train_y) * 100
    print("train:", accuracy)
    
    # Test Spam Classification
    test_data = sio.loadmat('./machine-learning-ex6/ex6/spamTest.mat')
    
    test_X = test_data['Xtest']
    test_y = test_data['ytest']
    
    m = len(test_X)
    predict = clf.predict(test_X).reshape((-1, 1))
    accuracy = np.mean(predict == test_y) * 100
    print("test:", accuracy)
    
    '''
    tmp = ['our', 'click', 'remov', 'guarante', 'visit', 'basenumb', 'dollar', 'will', 'price', 'pleas', 'nbsp',
           'most', 'lo', 'ga', 'dollarnumb']
    for i in range(len(tmp)):
        tmp[i] = int(vocabList[tmp[i]])
    print(tmp)
    '''
    
    topRankedWords(num_to_voca_list, clf.coef_[0])
    
    
main2_1()
```

```python
def isSpam(fileName, clf):
    vocabList, num_to_voca_list = getVocabList()
    
    email = readEmail(fileName)
    word_indices = processEmail(email)
    word_vector, features = emailFeatures(word_indices, vocabList)
    
    predict = clf.predict(features.reshape((1, -1)))
    
    print(predict == 1)

def main2_6():
    train_data = sio.loadmat('./machine-learning-ex6/ex6/spamTrain.mat')
    train_X = train_data['X']
    train_y = train_data['y']
    
    m = len(train_X)
    C = 1e-1
    clf = svm.LinearSVC(C=C)
    clf.fit(train_X, train_y.flatten())
    
    isSpam('./machine-learning-ex6/ex6/emailSample1.txt', clf)
    isSpam('./machine-learning-ex6/ex6/emailSample2.txt', clf)
    isSpam('./machine-learning-ex6/ex6/spamSample1.txt', clf)
    isSpam('./machine-learning-ex6/ex6/spamSample2.txt', clf)
    
main2_6()
```
